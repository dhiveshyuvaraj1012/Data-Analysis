```python
b={"team":["India","Srilanka","Pakistan"],"No. of time winners":[8,5,2]}
b1=pd.DataFrame(b)
b1
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the dictionary
df = pd.DataFrame(b)

# Set the index to the "team" column
df = df.set_index('team')

colors = ['blue', 'orangered', 'g']

ax = df['No. of time winners'].plot(kind='bar', color=colors)

for bar, color in zip(ax.containers, colors):
    ax.bar_label(bar, color=color)

# Add a title, axis labels, and remove legend
plt.title('Asia Cup Winners')
plt.xlabel('Team')
plt.ylabel('Number of Titles')
plt.legend().set_visible(False)

plt.show()
```
